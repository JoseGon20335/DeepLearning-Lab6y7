```python
import pandas as pd
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Load the datasets
books_df = pd.read_csv('data/Books.csv', low_memory=False)
ratings_df = pd.read_csv('data/Ratings.csv')
users_df = pd.read_csv('data/Users.csv')

# Check the shape of the datasets
print("Books dataset shape:", books_df.shape)
print("Ratings dataset shape:", ratings_df.shape)
print("Users dataset shape:", users_df.shape)
```

```python
users_df['Age'].fillna(users_df['Age'].mean(), inplace=True)

dataframe = pd.merge(ratings_df, books_df, on='ISBN')
dataframe = pd.merge(dataframe, users_df, on='User-ID')

dataframe.dropna(inplace=True)
dataframe.drop(['Image-URL-S', 'Image-URL-M', 'Image-URL-L', 'ISBN'], axis=1, inplace=True)

print("\nDataframe shape:", dataframe.shape)
print("\nDataframe missing values:\n", dataframe.isnull().sum())
print("\nDataframe duplicates:", dataframe.duplicated().sum())

dataframe.head()
```

```python
# CAst Age column to int without decimals
dataframe['Age'] = dataframe['Age'].astype(int)
dataframe['Year-Of-Publication'] = dataframe['Year-Of-Publication'].astype(int)

dataframe.head()
```

```python
print("Dataframe shape:", dataframe.shape)
print("\nDataframe missing values:\n", dataframe.isnull().sum())
```

```python
dataframe.to_csv('data/processed.csv', index=False)
```
